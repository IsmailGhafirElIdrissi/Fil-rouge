# MMA Fights Data Analysis
This notebook provides an analysis of the MMA fights dataset, focusing on cleaning the data by removing duplicates and filtering fights for the top 100 fighters based on their career stats.
### Key Tasks
- Data Loading
- Data Cleaning (Remove duplicates)
- Filter Fights for Top 100 Fighters
- Display Relevant Fights


```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the fights dataset
fights_df = pd.read_csv('../Data/Raw/Pro_MMA_Fights.csv')

# Display the first few rows of the dataset
fights_df.head()
```

## Data Cleaning
In this section, we will clean the data by removing any duplicate entries from the dataset.


```python
# Remove duplicates from the fights dataset
fights_df_clean = fights_df.drop_duplicates()

# Display the cleaned dataset
fights_df_clean.head()
```

## Filter Fights for Top 100 Fighters
Now, we will filter the fights to show only those involving the top 100 fighters based on their career stats. We will filter by fighter names, assuming the 'name' column exists in both datasets.


```python
# Load the top 100 fighters list from the previously created DataFrame (fighters_df)
top_100_fighters = pd.read_csv('../Data/Processed/Top_100_MMA_Fighters.csv')  # Assuming top 100 fighters DataFrame

# Filter fights that involve the top 100 fighters
fights_top_100 = fights_df_clean[
    (fights_df_clean['fighter1_name'].isin(top_100_fighters['fighter_name'])) |
    (fights_df_clean['fighter2_name'].isin(top_100_fighters['fighter_name']))
]

# Display the filtered dataset for the top 100 fighters
fights_top_100.head()
```

```python
# Save the filtered dataset for the top 100 fighters to a new CSV file
fights_top_100.to_csv('../Data/Processed/Fights_Top_100_Fighters.csv', index=False)
```

This analysis cleaned the MMA fights dataset by removing duplicates and filtered fights involving the top 100 fighters based on their career performance. The next steps could involve further analysis of these fights or predictive modeling based on these fights.

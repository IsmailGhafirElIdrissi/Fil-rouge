# MMA Data Analysis - Fighter Dataset
This notebook provides an analysis of the MMA fighter dataset, focusing on career performance, data cleaning, and classification of career-related columns. We will also create a top 100 fighters list based on their career stats.
### Key Tasks
- Data Loading
- Data Cleaning (Handle null values in specific columns)
- Classifying career columns (Pro MMA, Pro Boxing, etc.)
- Create Top 100 Fighters based on career stats


```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the fighters dataset
fighters_df = pd.read_csv('../Data/Raw/Pro_MMA_Fighters.csv')

# Display the first few rows of the dataset
fighters_df.head()
```

## Data Cleaning and Preparation
In this section, we will handle missing values in specific columns and classify career-related columns.


```python
# Handle missing values in the 'death_date' column by keeping rows with null values
fighters_df['death_date'] = fighters_df['death_date']  # No removal of null values for this column

# Define the career-related columns that need to be updated
career_columns = ['pro_mma', 'pro_boxing', 'pro_kickboxing', 'pro_muay_thai', 'pro_karate', 'pro_taekwondo', 'pro_wrestling', 'pro_bjj', 'pro_judo', 'pro_sambo']

# Replace values in all career columns
for col in career_columns:
    fighters_df[col] = fighters_df[col].replace({1: True, 0: False})
# Display the cleaned dataset with new classifications
fighters_df.head()
```

## Create Top 100 Fighters for Career Performance
We will now create a list of the top 100 fighters based on their career performance. For this example, we will consider a combination of win/loss ratio and number of fights.


```python
# Calculate win/loss ratio (assuming columns 'wins' and 'losses' exist)
fighters_df['win_loss_ratio'] = fighters_df['wins'] / (fighters_df['losses'] + 1)  # Adding 1 to avoid division by zero

# Sort the fighters by win_loss_ratio and select the top 100
top_100_fighters = fighters_df[['fighter_name', 'wins', 'losses', 'win_loss_ratio']].sort_values(by='win_loss_ratio', ascending=False).head(100)

# Display the top 100 fighters
top_100_fighters
```

```python
# Save the top 100 fighters to a new CSV file
top_100_fighters.to_csv('../Data/Processed/Top_100_MMA_Fighters.csv', index=False)

# Save the cleaned fighters dataset to a new CSV file
fighters_df.to_csv('../Data/Processed/Cleaned_MMA_Fighters.csv', index=False)
```
